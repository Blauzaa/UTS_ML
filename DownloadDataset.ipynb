```python

import yfinance as yf
import pandas as pd

ticker = "IDR=X"
start_date = "2019-01-01"
end_date = "2024-12-31"
nama_file_csv = "dataset_IDR_X_2019_2024.csv"
print(f"Mengunduh data {ticker} dari {start_date} hingga {end_date}...")
raw_data = yf.download(ticker, start=start_date, end=end_date)

if isinstance(raw_data.columns, pd.MultiIndex):
    raw_data.columns = raw_data.columns.get_level_values(0).str.lower()
else:
    raw_data.columns = raw_data.columns.str.lower()
raw_data.columns = raw_data.columns.str.replace(' ', '_')

numeric_columns = ['open', 'high', 'low', 'close', 'adj_close', 'volume']
for col in numeric_columns:
    if col in raw_data.columns:
        raw_data[col] = raw_data[col].astype(float)

raw_data.ffill(inplace=True)

raw_data.to_csv(nama_file_csv, index=True)

print(f"\nData berhasil diunduh, dibersihkan, dan disimpan sebagai '{nama_file_csv}'")
print("File ini sekarang 100% bersih dan siap digunakan.")

print("\nVerifikasi Tipe Data yang Disimpan:")
raw_data.info()
print("\nContoh Data yang Disimpan:")
display(raw_data.head())
```
